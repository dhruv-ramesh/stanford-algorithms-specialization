```python
import numpy as np 
import pandas as pd
from collections import deque 
import sys, threading
sys.setrecursionlimit(800000)
threading.stack_size(67108864)

nodes = 875714 + 1
graph = [[] for i in range(nodes)]
revgraph = [[] for i in range(nodes)]

file = open('hw5input.txt', 'r')

for line in file:

    ls = line.split(' ')
    graph[int(ls[0])] += [int(ls[1])]
    revgraph[int(ls[1])] += [int(ls[0])]

file.close()

visited = [False] * nodes
ftimes = [0] * nodes
ftimes2 = [0] * nodes
t = 0

vis2 = [False] * nodes
sccsize = [0] * nodes
leader = [0] * nodes
s = 0

def main():

    def dfs(g, n):

        global visited, ftimes, ftimes2, t

        visited[n] = True

        for v in g[n]:
            if not visited[v]:
                dfs(g, v)

        t += 1
        ftimes[t] = n
        ftimes2[n] = t


    def dfs_loop(g):

        global visited

        for n in range(1, nodes):
            if not visited[n]:
                dfs(g, n)

    dfs_loop(revgraph)
    
    torder = ftimes[::-1]
    
    def dfs2(g, n):
        global vis2, sccsize, s, leader
        
        vis2[n] = True
        leader[n] = s
        sccsize[s] += 1
        
        for v in g[n]:
            if not vis2[v]:
                dfs2(g, v)
    
    def dfs_loop2(g):
        
        global vis2, s
        
        for n in torder:
            if not vis2[n]:
                s = n
                dfs2(g, n)
     
    dfs_loop2(graph)
    
    print(sorted(sccsize)[::-1])
    
thread = threading.Thread(target=main)
thread.start()
```
